```python
beta = np.random.randn(2,1)
n = 1000
x = 2*np.random.rand(n,1)
y = 4+3*x+np.random.randn(n,1)

X = np.c_[np.ones((n,1)), x]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
lmbs = np.logspace(-8, 2, 15)
sgd = SGD_optimizer(fit_intercept = False, batch_size = 100, n_epochs = 1000,use_momentum= True, gamma = 0.5,regularization = 'l2', lmb = 0.001, lr=0.01)
mse_train, mse_test, r2_train,r2_test = evaluate_parameter(X_train,y_train, X_test, y_test, 'lambda', lmbs, sgd)
```
